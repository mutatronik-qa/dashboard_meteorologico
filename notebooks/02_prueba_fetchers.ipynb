# Prueba de Fetchers de Datos

Este notebook prueba los diferentes fetchers de datos meteorológicos.


```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve()))

from src.data_sources import OpenMeteoSource
from config import load_locations
```

```python
# Probar obtención de datos para múltiples ubicaciones
locations = load_locations()
source = OpenMeteoSource()

for loc in locations[:3]:  # Primeras 3 ubicaciones
    try:
        data = source.get_current_weather(loc.lat, loc.lon)
        print(f"{loc.name}: {data.get('temperature')}°C")
    except Exception as e:
        print(f"Error en {loc.name}: {e}")
```
