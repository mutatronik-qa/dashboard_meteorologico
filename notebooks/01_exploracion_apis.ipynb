# Exploración de APIs Meteorológicas

Este notebook explora las diferentes APIs meteorológicas disponibles.


```python
import sys
from pathlib import Path

# Agregar src al path
sys.path.append(str(Path('..').resolve()))

from src.data_sources import OpenMeteoSource
from config import get_settings, load_locations
```

```python
# Cargar configuración
settings = get_settings()
locations = load_locations()

print(f"Ubicaciones cargadas: {len(locations)}")
for loc in locations:
    print(f"  - {loc.name}: ({loc.lat}, {loc.lon})")
```

```python
# Probar Open-Meteo (sin API key)
openmeteo = OpenMeteoSource()

medellin = locations[0]  # Medellín
data = openmeteo.get_current_weather(medellin.lat, medellin.lon)

print(f"Datos de Open-Meteo:")
print(f"  Temperatura: {data.get('temperature')}°C")
print(f"  Humedad: {data.get('humidity')}%")
print(f"  Precipitación: {data.get('precipitation')} mm")
```
