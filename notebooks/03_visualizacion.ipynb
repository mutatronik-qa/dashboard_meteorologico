# Visualización de Datos Meteorológicos

Este notebook muestra cómo crear visualizaciones con los datos obtenidos.


```python
import sys
from pathlib import Path
sys.path.append(str(Path('..').resolve()))

from src.data_sources import OpenMeteoSource
from src.visualizers.plots import create_comparison_chart
from config import load_locations
```

```python
# Obtener datos y crear visualizaciones
locations = load_locations()
source = OpenMeteoSource()

data_list = []
for loc in locations[:5]:
    data = source.get_current_weather(loc.lat, loc.lon)
    data['location'] = {'name': loc.name, 'lat': loc.lat, 'lon': loc.lon}
    data_list.append(data)

# Crear gráfico de comparación
fig = create_comparison_chart(data_list, metric='temperature')
fig.show()
```
